<br id="correlations">
# 5. Correlations

Practiced making a good heatmap for the data. The final heatmap has the numerical correlations on each square, along with its color indicating correlation. Also, since heatmaps generally display the same data twice, I removed the top half to create a triangle shaped heatmap. 

```python
# Calculate correlations between numeric features
correlations = df.corr()
correlations
```

```python
# Make the figsize 7 x 6
plt.figure(figsize=(7,6))
# Plot heatmap of correlations
sns.heatmap(correlations)
plt.show()
```

```python
# Change color scheme
sns.set_style("white")
# Make the figsize 10 x 8
plt.figure(figsize=(10,8))

# Plot heatmap of correlations
sns.heatmap(correlations)
plt.show()
```

```python
# Make the figsize 10 x 8
plt.figure(figsize=(10,8))
# Plot heatmap of annotated correlations
sns.heatmap(correlations*100, annot=True, fmt = '.0f')
plt.show()
```

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(correlations, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
```

```python
# Make the figsize 10 x 8
plt.figure(figsize=(10,8))
# Plot heatmap of correlations
sns.heatmap(correlations*100, annot=True, fmt = '.01f',mask = mask)
plt.show()
```

```python
# Make the figsize 9 x 8
plt.figure(figsize=(9,8))
# Plot heatmap of correlations
sns.heatmap(correlations*100,annot=True, fmt = '.01f', mask=mask, cbar=False)
plt.show()
```
