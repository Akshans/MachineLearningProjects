```python
qarot_data = pd.read_csv('QAROT_data_use.csv') #sanctionstrenght. placementstrength categorical
qarot_managed = pd.read_excel('QAROT_data_managed.xlsx')
```

```python
qarot_country = qarot_managed[['country']]
qarot_country
```

```python
qarot_data = qarot_data[qarot_data.columns.drop(list(qarot_data.filter(regex='Unnamed')))]
```

```python
qarot_data['adopted.reform.flag'] = qarot_data['adopted.reform.flag'].fillna(0)
qarot_data['implemented.reform.flag'] = qarot_data['implemented.reform.flag'].fillna(0)
qarot_data['quota.reform.number'] = qarot_data['quota.reform.number'].fillna(0)
qarot_data['implement.reform.number'] = qarot_data['implement.reform.number'].fillna(0)
qarot_data['percent.stated.threshold'] = qarot_data['percent.stated.threshold'].fillna(0)
qarot_data['percent.of.seats'] = qarot_data['percent.of.seats'].fillna(0)
qarot_data['defacto.threshold'] = qarot_data['defacto.threshold'].fillna(0)
qarot_data['sanctions'] = qarot_data['sanctions'].fillna(0)
qarot_data['placement'] = qarot_data['placement'].fillna(0)
qarot_data['reserved.separate.tier'] = qarot_data['reserved.separate.tier'].fillna(0)
qarot_data['reserved.best.loser'] = qarot_data['reserved.best.loser'].fillna(0)
qarot_data['reserved.indirect'] = qarot_data['reserved.indirect'].fillna(0)
qarot_data['reserved.unspecified'] = qarot_data['reserved.unspecified'].fillna(0)
qarot_data['effective.quota'] = qarot_data['effective.quota'].fillna(0)
qarot_data['women.rep'] = qarot_data['women.rep'].fillna(0)
```
